# 8 . All Features Vs Quality Comparison

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named 'data'
features = [col for col in data.columns if col != 'quality']

num_rows = 4
num_cols = 3
fig, axes = plt.subplots(num_rows, num_cols, figsize=(16, 12))

for i, feature in enumerate(features):
    row = i // num_cols
    col = i % num_cols

    # Plot with 'quality' as hue to assign the palette and disable dodge for alignment
    sns.barplot(x='quality', y=feature, hue='quality', data=data, ax=axes[row, col],
                palette=sns.color_palette("husl", len(data['quality'].unique())), dodge=False)
    
    axes[row, col].set_title(f'Quality vs {feature}')
    axes[row, col].set_xlabel('Quality')
    axes[row, col].set_ylabel(feature)
    
    # Remove legend if it exists
    if axes[row, col].get_legend() is not None:
        axes[row, col].get_legend().remove()
    
    # Rotate x-axis tick labels to prevent overlap
    axes[row, col].tick_params(axis='x', rotation=35)

plt.tight_layout()
plt.show()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA

# Create a pipeline that applies scaling, PCA, and then the RandomForestClassifier
pipeline = Pipeline([
    ('scaler', st),   # assuming 'st' is your fitted StandardScaler
    ('pca', pca),     # assuming 'pca' is your fitted PCA transformer
    ('classifier', rf)
])

# Save the entire pipeline to a file
joblib.dump(pipeline, 'wine_quality_pipeline.pkl')

# To load the pipeline later:
loaded_pipeline = joblib.load('wine_quality_pipeline.pkl')

# Predict on new data using the loaded pipeline
test = loaded_pipeline.predict(new_data)  # new_data should be in raw feature form
if test[0] == 1: 
    print("Good Quality Wine")
else:
    print("Bad Quality Wine")
```
